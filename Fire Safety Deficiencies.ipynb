### A summary that shows Hazard Area Deficiencies for all states from 2013-2019

```python
#print(df['Deficiency Category']. unique())
```

```python
#sd = df[df['Deficiency Category'] == 'Smoke Deficiencies']
#sd.head(5)
```

```python
df_smoke = df[df['Deficiency Category'] == 'Hazardous Area Deficiencies']

plt.hist(df_smoke['Provider State'], histtype='bar', ec='black')
df_smoke = df_smoke.sort_values('Deficiency Category', ascending=True)
plt.xticks(rotation=90)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 6}
plt.xlabel('Locations')
plt.ylabel('Frequency')
plt.title('Hazardous Area Deficiencies Summary')
plt.show()
```
